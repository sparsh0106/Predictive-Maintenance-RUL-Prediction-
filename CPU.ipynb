```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
df['year'] = df['timestamp'].dt.year
df['month'] = df['timestamp'].dt.month
df['day'] = df['timestamp'].dt.day
df['hour'] = df['timestamp'].dt.hour
df['minute'] = df['timestamp'].dt.minute
df['second'] = df['timestamp'].dt.second
df = df.drop('timestamp', axis=1)
```

```python
df.drop(columns = ['year','second'], axis=1, inplace=True)
```

```python
X = df.drop('rul', axis=1)
y = df['rul']
```

```python
from sklearn.preprocessing import StandardScaler

SS = StandardScaler()
X_scaled = SS.fit_transform(X)
X = pd.DataFrame(X_scaled, columns=X.columns)
```

```python
X.head()
```

```python
# from sklearn.preprocessing import MinMaxScaler
#
# mms = MinMaxScaler()
# df['rul'] = mms.fit_transform(df[['rul']])
# df['rul'].head()
```

```python
y_scaled = df['rul']
```

```python
import numpy as np
from sklearn.decomposition import PCA

pca = PCA()
pca.fit(X_scaled)
cumsum = np.cumsum(pca.explained_variance_ratio_)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (15,15))
plt.plot(range(1,len(cumsum) + 1), cumsum, marker = 'o', color = 'r')
plt.xlabel('Number of components')
plt.ylabel('Cumulative explained variance')
plt.title('Choosing the number of components')
plt.grid('True')
plt.show()
```

```python
import time
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
pca = PCA(n_components = 50)
X_scaled_pca = pca.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled_pca, y_scaled, test_size = 0.2, random_state = 42)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
```

```python
start = time.time()
model.fit(X_train, y_train)
end = time.time()
print(f"Training time: {end - start} seconds")
```

```python
r2 = r2_score(y_test, model.predict(X_test))
mae = mean_absolute_error(y_test, model.predict(X_test))
mse = mean_squared_error(y_test, model.predict(X_test))
rmse = np.sqrt(mse)
```

```python
print(f"Accuracy: {r2*100:.2f}%")
print(f"MAE: {mae*100:.2f}%")
print(f"MSE: {mse*100:.2f}%")
print(f"RMSE: {rmse*100:.2f}%")
```
